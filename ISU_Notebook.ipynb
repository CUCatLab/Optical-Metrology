$HalfWaveplate$

```python
x =-13

def HWP(x):
    hwp = np.array([np.cos(x)**2 - np.sin(x)**2, 2*np.cos(x)*np.sin(x)], [2*np.cos(x)*np.sin(x), np.sin(x)**2 - np.cos(x)**2]) #half waveplate matrix

    return(hwp)
```

$Horizontal Polarizer With Leakage$

```python
def HPL(n=.1):
    hp = np.array([[np.sqrt(1-n**2), 0],[0, n]]) #horizontal polarizer
    return(hp)
```

$Polarizaer setup with 2 HWP and 1 PBS$

```python
def pol(x,n=.1):
    p  = HWP.transpose @ HPL @ HWP
    return(p)
print(pol)
```
